## 3.1 Linear
Each action logit is a linear combination of pixels and parameters: 

$$logit_{left} = \sum_i w_i x_i + b_i$$

The problem is fairly simple, and a linear policy should be able to detect whether there is pavement in front of the car or not, also whether pixels on left/right has grass on it. 
In principle we only need four pixels that can weigh whether it is useful to go forward/backward/left/right (depending on those four values), and a linear policy could map into actions. Whether it will work in practice is of course difficult to say.

A linear policy would probably not be optimal. Most likely it can only learn simple "local" linear rules, and will be able to slowly drive forward if there is pavement in that direction. A linear model cannot model interdepencies between pixels, and doing high speed turns etc would be harder.

A fixed camera would create less linearity to the observation space and it would work worse. If you use bird eye view and make the observations rotate according to car direction, a linear model would break down (as a pixel would no longer be a pixel).
